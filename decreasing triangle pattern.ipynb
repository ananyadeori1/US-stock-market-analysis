```python
djia = yf.Ticker("DJIA").history('2y')
```

```python
mpf.plot(djia, savefig = 'img.png')
```

```python
mpf.plot(djia['2020-1-1':'2020-04-30'])
```

```python
djia.to_csv('djia.csv')
data=pd.read_csv('djia.csv')
data.info()
```

```python

data.Date=pd.to_datetime(data.Date)  #change 
data=data.set_index('Date')
```

```python
mpf.plot(data)
```

```python
mpf.plot(data['2020'])
```

```python
mpf.plot(data['2020'], type='line')
```

```python
mpf.plot(data['2020'], type='candle')
```

```python
mpf.plot(data['2020'], type='renko')
```

```python
mpf.plot(data, type='pnf')
```
